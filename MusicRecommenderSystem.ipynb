```python
#importing dependencies
import pandas as pd
import recom_class as recon
```

```python
#loading the datasets
song_ID = pd.read_csv('songUserID.csv')
song_Data = pd.read_csv('songData.csv')
```

```python
song_ID.head()
```

```python
song_Data.head()
```

```python
#creating new dataframe by merging the given two dataframes
Songs = pd.merge(song_ID, song_Data.drop_duplicates(['song_id']), on='song_id', how='left')
Songs.head()
```

```python
#merging the features 'title' and 'artist' to create a new feature 'song' for concise handling of the dataset
Songs['Song'] = Songs['title']+'-'+Songs['artist_name']
Songs.head()
```

```python
#limiting training entries to 50000 for faster prediction (more the input, greater is the time to predict)
Songs = Songs.head(50000)
```

```python
#aggregating the listen count of individual songs into a single value
Grouped_count = Songs.groupby(['Song']).agg({'listen_count':'count'}).reset_index()
Grouped_count.head(10)
```

```python
#finding total listen count of all the songs in the dataframe
Sum = Grouped_count['listen_count'].sum()

#creating new feature percentile
Grouped_count['Percentage'] = (Grouped_count['listen_count'] / Sum ) * 100
print("\n         SHOWING THE MOST POPULAR SONGS IN THE DATASET\n")
Grouped_count.sort_values(['listen_count', 'Song'], ascending=[0,1])
```

```python
#creating a instance to find similar items
similarItems = recon.Similar_Items_Recommender()
similarItems.construct(Songs, 'user_id', 'Song')
```
